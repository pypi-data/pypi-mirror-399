# 欢迎使用 BulletTrade 研究环境

- 默认根目录：`~/bullet-trade`（Windows: `~\\bullet-trade`）
- 设置文件：`~/.bullet-trade/setting.json`，可调整 `host/port/root_dir/env_path/open_browser` 等
- `.env` 路径：默认使用根目录下的 `.env`，不会读取当前工作目录的 `.env`
- 首次启动已复制示例 Notebook（不覆盖已有文件），并写入代码片段供命令面板插入

想要快速开始？尝试下面的示例 Notebook 或运行回测/实盘命令：
```bash
bullet-trade backtest strategies/demo_strategy.py --start 2025-01-01 --end 2025-06-01
bullet-trade live strategies/demo_strategy.py --broker qmt
```


## 小贴士
- 如果需要公网访问，请在 `setting.json` 启用密码或证书，避免使用裸露的 `0.0.0.0`。
- 想换根目录？编辑 `root_dir` 并同步修改 `env_path` 即可。
- 代码片段可在 Command Palette 搜索 "BulletTrade" 后插入。

```python
# カスタムパラメータを指定する例
import pandas as pd
import os
from jphrase import PhraseExtracter

path = os.path.dirname(os.path.abspath("__file__")) 
df_texts = pd.read_table(path + "/data/text.tsv", header=None, lineterminator='\n', names=["sentence"])

# パラメータをカスタマイズ
extractor = PhraseExtracter(
    verbose=1,
    size_sentence=5000,  # 一回で処理するセンテンスの数
    min_count=6,  # 文字連カウントの最小数閾値
    max_length=16,  # 文字の長さ最大値
    min_length=4,  # 文字の長さ最小値
    weight_freq=1,  # 頻度への重み（たくさんある連なりを重視）
    weight_len=1,  # 長さへの重み（長い連なりを重視）
    threshold_originality=0.50,  # 独自性の閾値（0.0なら絞らず、0.9なら似たフレーズが除去される）
    knowns=[]  # 既知語リスト（優先的に検出したい語句）
)

df_result = extractor.get_dfphrase(df_texts["sentence"])
display(df_result)
```
